```python
#---- Dibujando los rasters de neuronas -------
import csv
import numpy as np
import matplotlib.pyplot as plt
import Caro_lectura_archivo as lectArchivo
```

```python
vacios,neu_dpc1 = lectArchivo.lectura_archivo('Practica1_Estadistica/TiemposNeuDPC1.csv')
```

```python
len(neu_dpc1)
```

```python
def indices_reales(vacios):
    indices_final = []
    for i in range(len(vacios)):
        indices_corregidos = np.array(vacios)[i] - vacios.index(vacios[i])
        indices_final.append(indices_corregidos)
    return np.array(indices_final)
```

```python
indices_vac = indices_reales(vacios)

fig,axs=plt.subplots(4,1, sharex='all', figsize=[15,10])
fig.suptitle('Raster de la tarea de discriminacin de patrones de la neurona DPC1', fontsize=20, fontweight=500)
axs[0].eventplot(neu_dpc1[:indices_vac[0]],linelengths=.5, colors='#bf1140')
axs[0].set_title('Clase 1',size=15)

axs[1].eventplot(neu_dpc1[indices_vac[0]:indices_vac[1]],linelengths=.5, color='#f2285f')
axs[1].set_title('Clase 2',size=15)

axs[2].eventplot(neu_dpc1[indices_vac[1]:indices_vac[2]],linelengths=.5, colors='#f34c79')
axs[2].set_title('Clase 3',size=15)


axs[3].eventplot(neu_dpc1[indices_vac[2]:],linelengths=.5, colors='#f57698')
axs[3].set_title('Clase 4',size=15)

plt.xlabel('Tiempo[s]',size=13)  
plt.show()
```

```python
def raster_plot(nombre_archivo,archivo,vacios,colores):
    indices_vac = indices_reales(vacios)

    fig,axs=plt.subplots(4,1, sharex='all', figsize=[15,10])
    fig.suptitle(f'Raster tarea de discriminacin de patrones de la neurona {nombre_archivo}', fontsize=20, fontweight=500)
    axs[0].eventplot(archivo[:indices_vac[0]],linelengths=0.5, colors=colores[0])
    axs[0].set_title('Clase 1',size=15)

    axs[1].eventplot(archivo[indices_vac[0]:indices_vac[1]],linelengths=.5, color=colores[1])
    axs[1].set_title('Clase 2',size=15)

    axs[2].eventplot(archivo[indices_vac[1]:indices_vac[2]],linelengths=.5, colors=colores[2])
    axs[2].set_title('Clase 3',size=15)

    axs[3].eventplot(archivo[indices_vac[2]:],linelengths=.5, colors=colores[3])
    axs[3].set_title('Clase 4',size=15)

    plt.xlabel('Tiempo[s]',size=13) 
    
    plt.show()
```

```python
vacios,neu_dpc2 = lectArchivo.lectura_archivo('Practica1_Estadistica/TiemposNeuDPC2.csv')
colores = ['#094d80','#146bab','#3aa8fa','#8bc8f6']
raster_plot('DPC2',neu_dpc2,vacios,colores)
```

```python
def raster_det(nombre_archivo,archivo,vacios,colores):
    indices_vac = indices_reales(vacios)
    archivo = [archivo[i][(archivo[i]<=3.5) & (archivo[i]>=-2)] for i in range(len(archivo))]
    
    fig,axs=plt.subplots(6,1, sharex='all', figsize=[15,18])
    fig.suptitle(f'Raster tarea de deteccin de amplitudes de la neurona {nombre_archivo}', fontsize=20, fontweight=500)
    axs[0].eventplot(archivo[:indices_vac[0]],linelengths=0.5, colors=colores[0])
    axs[0].set_title('Bloque 1: 0m',fontsize=13)

    axs[1].eventplot(archivo[indices_vac[0]:indices_vac[1]],linelengths=.5, color=colores[1])
    axs[1].set_title('Bloque 2: 6m',fontsize=13)

    axs[2].eventplot(archivo[indices_vac[1]:indices_vac[2]],linelengths=.5, colors=colores[2])
    axs[2].set_title('Bloque 3: 8m',fontsize=13)

    axs[3].eventplot(archivo[indices_vac[2]:indices_vac[3]],linelengths=.5, colors=colores[3])
    axs[3].set_title('Bloque 4: 10m',fontsize=13)
    
    axs[4].eventplot(archivo[indices_vac[3]:indices_vac[4]],linelengths=.5, colors=colores[4])
    axs[4].set_title('Bloque 5: 12m',fontsize=13)
    
    axs[5].eventplot(archivo[indices_vac[4]:],linelengths=.5, colors=colores[5])
    axs[5].set_title('Bloque 6: 24m',fontsize=13)
   
  
    
    plt.xlabel('Tiempo[s]',fontsize=13) 
    plt.show()
```

```python
vacios,det_s1a = lectArchivo.lectura_archivo('Practica2_Estadistica/Neu_Det_S1A.csv')
colores = ['#2e084c','#480c77','#661aa2','#9934e8','#af5eef','#b08bcd']
raster_det('S1_A',det_s1a,vacios,colores)
```
